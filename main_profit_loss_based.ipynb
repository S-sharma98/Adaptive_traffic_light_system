```python
import os
import pprint as pp
import csv
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import cvlib as cv 
import numpy as np
from cvlib.object_detection import draw_bbox
```

```python
input_path =f'./gdrive/My Drive/model'
output_path  = f'./gdrive/My Drive/output_images'
```

```python
# imp variables
cnt = 1
cnt1 = 0
m = int(70)
# minimum limit of maximum time
T = int(input())

def profit_time(T , len_green_time):
    p = T - len_green_time
    return p

def calculate_timings(vehicle_count , Maximum_time):
    # time of green signal = t
    y = 3 * vehicle_count
    if vehicle_count < 5:
        t = 10
    else:
        if y > Maximum_time:
            t= Maximum_time
        else:
            t = y
    return t        
    

for filename in os.listdir(input_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        N = 0
        car = 0
        bus = 0
        truck = 0
        bike = 0
        img_file=os.path.join(input_path, filename) 
        original_img = cv2.imread(img_file)
        original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
        img=cv2.resize(original_img,(416,416))
        
        bbox, label, conf = cv.detect_common_objects(img)
        for element in label:
            if element == "car":
                car = car+1
            elif element == "bus":
                bus = bus +1
            elif element == "truck":
                truck = truck + 1
            elif element == "motorcycle"
                bike = bike+1
        N = car + bus + truck
        
        p = profit_time(T, m)
        x = cnt - cnt1
        if x ==1:
            n = calculate_timings(N , T+p)
            t1 = n
            t2 = n
            t3 = 2*T
            t4 = 3*T
            print("len1 - green - " , t1)
            print("len2 - red - " , t2)
            print("len3 - red - " , t3)
            print("len4 - red - " , t4)
        m = n
        p = profit_time(T, m)
        cnt += 1
        print("\n")
    
    if x == 2:
        n = calculate_timings(N , T+p)
        t1 = 3*T
        t2 = n
        t3 = n
        t4 = (3*T - m)
        print("len1 - red - " , t1)
        print("len2 - green - " , t2)
        print("len3 - red - " , t3)
        print("len4 - red - " , t4)
    
    
        m = n
        p = profit_time(T, m)
        cnt += 1
        print("\n")
    
    if x == 3:
        n = calculate_timings(N , T+p)
        t1 = (3*T - m)
        t2 = 3*T
        t3 = n
        t4 = n
            
        print("len1 - red - " , t1)
        print("len2 - red - " , t2)
        print("len3 - green - " , t3)
        print("len4 - red - " , t4)
    
        
        m = n
        p = profit_time(T, m)
        cnt += 1
        print("\n")
    
    if x == 4:
        n = calculate_timings(N , T+p)
        t1 = n
        t2 = (3*T - m)
        t3 = 3*T
        t4 = n
        
        print("len1 - red - " , t1)
        print("len2 - red - " , t2)
        print("len3 - red - " , t3)
        print("len4 - green - " , t4)
            
        cnt= 1
        print("\n")
            
    
    else:
        continue
```
